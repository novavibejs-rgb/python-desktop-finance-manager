```python
import sqlite3
```

```python
conn = sqlite3.connect("financeiro.db")
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS socios (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nome TEXT NOT NULL
)
""")
```

```python
cursor.execute("DELETE FROM socios")
```

```python
conn.commit()
conn.close()
```

```python
def cadastrar_socio(nome):
    conn = sqlite3.connect("financeiro.db")
    cursor = conn.cursor()

    cursor.execute("""
    INSERT INTO socios (nome)
    VALUES (?)
    """, (nome,))

    conn.commit()
    conn.close()
```

```python
def listar_socios():
    conn = sqlite3.connect("financeiro.db")
    cursor = conn.cursor()

    cursor.execute("""
    SELECT * FROM socios
    """)

    socios = cursor.fetchall()

    conn.close()

    return socios
```

```python
def atualizar_socio(id_socio, novo_nome):
    conn = sqlite3.connect("financeiro.db")
    cursor = conn.cursor()

    cursor.execute("""
    UPDATE socios
    SET nome = ?
    WHERE id = ?
    """, (novo_nome, id_socio))

    conn.commit()
    conn.close()
```

```python
def excluir_socio(id_socio):
    conn = sqlite3.connect("financeiro.db")
    cursor = conn.cursor()

    cursor.execute("""
    DELETE FROM socios
    WHERE id = ?
    """, (id_socio,))

    conn.commit()
    conn.close()
```

```python
cadastrar_socio("leticia")
```

```python
print(listar_socios())
```

```python
atualizar_socio(6,"eliane")
```

```python
excluir_socio(5)
```
