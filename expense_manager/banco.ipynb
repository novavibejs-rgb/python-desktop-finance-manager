```python
import sqlite3
from datetime import datetime
```

```python
conn = sqlite3.connect("financeiro.db")
cursor = conn.cursor()
```

```python
#criando tabelas

cursor.execute("""
CREATE TABLE IF NOT EXISTS socios (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nome TEXT NOT NULL
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS servicos (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    cliente TEXT NOT NULL,
    descricao TEXT NOT NULL,
    valor REAL NOT NULL,
    forma_pagamento TEXT,
    data DATETIME DEFAULT CURRENT_TIMESTAMP
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS vales (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nome_socio TEXT NOT NULL,
    valor REAL NOT NULL,
    descricao TEXT,
    data TEXT
)
""")

conn.commit()
```

```python
#commit
conn.commit()
conn.close()
```

funçao de socio

```python
def cadastrar_socio(nome):
    conn = sqlite3.connect("financeiro.db")
    cursor = conn.cursor()

    cursor.execute("""
    INSERT INTO socios (nome)
    VALUES (?)
    """, (nome,))

    conn.commit()
    conn.close()
```

```python
def listar_socios():
    conn = sqlite3.connect("financeiro.db")
    cursor = conn.cursor()

    cursor.execute("""
    SELECT * FROM socios
    """)

    socios = cursor.fetchall()

    conn.close()

    return socios
```

```python
def atualizar_socio(id_socio, novo_nome):
    conn = sqlite3.connect("financeiro.db")
    cursor = conn.cursor()

    cursor.execute("""
    UPDATE socios
    SET nome = ?
    WHERE id = ?
    """, (novo_nome, id_socio))

    conn.commit()
    conn.close()
```

```python
def excluir_socio(id_socio):
    conn = sqlite3.connect("financeiro.db")
    cursor = conn.cursor()

    cursor.execute("""
    DELETE FROM socios
    WHERE id = ?
    """, (id_socio,))

    conn.commit()
    conn.close()
```

funçao de serviço

```python
def faturamento_por_mes(mes_ano):
    cursor.execute("""
    SELECT SUM(valor)
    FROM servicos
    WHERE data LIKE ?
    """, (f"%{mes_ano}%",))

    total = cursor.fetchone()[0]
    return total if total is not None else 0
```

```python
def calcular_faturamento_total():
    cursor.execute("""
    SELECT SUM(valor) FROM servicos
    """)

    total = cursor.fetchone()[0]

    # Se não houver serviços, retorna 0
    return total if total is not None else 0
```

```python
def listar_servicos():
    cursor.execute("""
    SELECT * FROM servicos
    """)

    return cursor.fetchall()
```

```python
def buscar_servico_por_id(id_servico):
    cursor.execute("""
    SELECT * FROM servicos
    WHERE id = ?
    """, (id_servico,))

    return cursor.fetchone()
```

```python
def adicionar_servico(nome_cliente, descricao, valor, forma_pagamento):
    conn = sqlite3.connect("financeiro.db")
    cursor = conn.cursor()
    data = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    cursor.execute("""
    INSERT INTO servicos
    (cliente, descricao, valor, forma_pagamento, data)
    VALUES (?, ?, ?, ?, ?)
    """, (nome_cliente, descricao, valor, forma_pagamento, data))

    conn.commit()
    conn.close()
```

```python
def atualizar_servico(id_servico, nome_cliente, descricao, valor, forma_pagamento):
    cursor.execute("""
    UPDATE servicos
    SET cliente = ?,
        descricao = ?,
        valor = ?,
        forma_pagamento = ?
    WHERE id = ?
    """, (nome_cliente, descricao, valor, forma_pagamento, id_servico))

    conn.commit()
```

```python
def excluir_servico(id_servico):
    cursor.execute("""
    DELETE FROM servicos
    WHERE id = ?
    """, (id_servico,))

    conn.commit()
```

```python
def faturamento_semana():
    sete_dias = datetime.now() - timedelta(days=7)
    sete_dias = sete_dias.strftime("%Y-%m-%d %H:%M:%S")

    cursor.execute("""
    SELECT SUM(valor)
    FROM servicos
    WHERE data >= ?
    """, (sete_dias,))

    total = cursor.fetchone()[0]
    return total if total is not None else 0
```
