```python
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from tkinter import filedialog
from PIL import Image
from matplotlib.figure import Figure
import customtkinter as ctk
from datetime import datetime

class App:

    def __init__(self):
        self.root = ctk.CTk()
        self.root.title("Sistema Financeiro",)
        self.root.geometry("800x520")
        self.root.resizable(False, False)
        self.root.configure(fg_color="#0E36A3")
    

        # ===== CONFIG GRID PRINCIPAL =====
        self.root.grid_rowconfigure(0, weight=0)  # topo
        self.root.grid_rowconfigure(1, weight=1)  # conteúdo

        self.root.grid_columnconfigure(0, minsize=200)
        self.root.grid_columnconfigure(1, weight=1)  # dashboard

        # ===== TOP BAR (RELÓGIO) =====
        self.top_frame = ctk.CTkFrame(self.root, fg_color="#686666")
        self.top_frame.grid(row=0, column=0, columnspan=2, sticky="ew", padx=10, pady=10)

        self.relogio_label = ctk.CTkLabel(
            self.top_frame,
            text="relogio",
            font=("Segoe UI Emoji", 24,),
            text_color="#F8F5F5"
        )
        self.relogio_label.pack(pady=10)

        # ===== MENU LATERAL =====
        self.menu_frame = ctk.CTkFrame(self.root, fg_color="#686666",width=250)
        self.menu_frame.grid(row=1, column=0, sticky="nsew", padx=10, pady=10)
       
        ctk.CTkLabel(
            self.menu_frame,
            text="MENU",
            font=("Arial", 30, "bold"),
            text_color="#0A0404"
        ).pack(pady=5)

        self.menu_frame.grid_propagate(False)

        #buttons do menu
        ctk.CTkButton(self.menu_frame, text="Inicio",border_width=2,border_color="#0D0E0D",text_color="#0D0E0D",fg_color="#0E36A3",command=self.mostrar_inicio).pack(pady=20)
        ctk.CTkButton(self.menu_frame, text="Perfil", border_width=2,border_color="#0A0A0A",text_color="#0A0A0A", fg_color="#0E36A3",command=self.mostrar_perfil).pack(pady=20)
        ctk.CTkButton(self.menu_frame, text="Adicionar serviço", border_width=2,border_color="#0A0A0A",text_color="#0A0A0A", fg_color="#0E36A3", command=self.mostrar_adicionar_servico).pack(pady=20)
        ctk.CTkButton(self.menu_frame, text="Fazer vale", border_width=2,border_color="#0A0A0A",text_color="#0A0A0A", fg_color="#0E36A3", command=self.mostrar_fazer_vale).pack(pady=20)

        self.content_frame = ctk.CTkFrame(self.root)
        self.content_frame.grid(row=1, column=1, sticky="nsew", padx=10, pady=10)
        self.content_frame.grid_rowconfigure(0, weight=1)
        self.content_frame.grid_columnconfigure(0, weight=1)

       # ===== FRAMES DAS TELAS =====
        self.frame_inicio = ctk.CTkFrame(self.content_frame, fg_color="#686666")
        self.frame_perfil = ctk.CTkFrame(self.content_frame, fg_color="#686666")
        self.frame_servico = ctk.CTkFrame(self.content_frame, fg_color="#686666")
        self.frame_vale = ctk.CTkFrame(self.content_frame, fg_color="#686666")

        # Todos ocupam o mesmo lugar
        for frame in (
            self.frame_inicio,
            self.frame_perfil,
            self.frame_servico,
            self.frame_vale
        ):
            frame.grid(row=0, column=0, sticky="nsew")
        
       # Criar todas as telas uma única vez
        self.criar_inicio()
        self.criar_perfil()
        self.criar_servico()
        self.criar_vale()

        # Mostrar a tela inicial
        self.frame_inicio.tkraise()

        self.atualizar_relogio()

        self.root.mainloop()

    # ===== RELÓGIO =====
    def atualizar_relogio(self):
        agora = datetime.now().strftime(" %d/%m/%Y  |  %H:%M:%S")
        self.relogio_label.configure(text=agora)
        self.root.after(1000, self.atualizar_relogio)

    def mostrar_adicionar_servico(self):
        self.frame_servico.tkraise()
    

    def mostrar_fazer_vale(self):
        self.frame_vale.tkraise()
        

    def mostrar_perfil(self):
        self.frame_perfil.tkraise()
        
    
    def mostrar_inicio(self):
        self.frame_inicio.tkraise()


    def criar_inicio(self):
        self.rd_pe = ctk.CTkFrame(self.frame_inicio, fg_color="#20B960", height=100)
        self.rd_pe.grid(row=0, column=0, columnspan=3, sticky="nsew", padx=10, pady=10)

        ctk.CTkLabel(
            self.rd_pe,
            text="💰 Saldo Total",
            font=("Arial", 25, "bold"),
            text_color="#0D0C0E"
        ).grid(row=0, column=0, padx=10, pady=10)

        self.lbl_saldo_total = ctk.CTkLabel(
            self.rd_pe,
            text="R$ 1.800,00",
            font=("Arial", 25, "bold"),
            text_color="#FCF8F8"
        )
        self.lbl_saldo_total.grid(row=1, column=0, padx=10)

        ctk.CTkLabel(
            self.rd_pe,
            text="📥 Reserva",
            font=("Arial", 25, "bold"),
            text_color="#0D0C0E"
        ).grid(row=0, column=1, padx=10)

        self.lbl_reserva = ctk.CTkLabel(
            self.rd_pe,
            text="R$ 0,00",
            font=("Arial", 25, "bold"),
            text_color="#FCF8F8"
        )
        self.lbl_reserva.grid(row=1, column=1)

        ctk.CTkLabel(
            self.rd_pe,
            text="📈 Lucro Líquido",
            font=("Arial", 25, "bold"),
            text_color="#0D0C0E"
        ).grid(row=0, column=2, padx=10)

        self.lbl_lucro = ctk.CTkLabel(
            self.rd_pe,
            text="R$ 0,00",
            font=("Arial", 25, "bold"),
            text_color="#FCF8F8"
        )
        self.lbl_lucro.grid(row=1, column=2)

        # gráfico
        self.fig = Figure(figsize=(5, 3), dpi=100)
        self.ax = self.fig.add_subplot(111)
        self.ax.set_title("Desempenho Mensal")
        self.ax.plot([1, 2, 3, 4], [10, 25, 15, 30])

        self.canvas = FigureCanvasTkAgg(self.fig, master=self.frame_inicio)
        self.canvas.draw()
        self.canvas.get_tk_widget().grid(row=1, column=0, columnspan=3, pady=10)
            

    def criar_perfil(self):

        ctk.CTkLabel(
                self.frame_perfil,
                text="Perfil do Usuário",
                font=("Arial", 30, "bold"),
                text_color="#0D0C0E"
            ).pack(pady=10)

        self.card1 = ctk.CTkFrame(self.frame_perfil, width=200, height=300, fg_color="#20B960", corner_radius=10)
        self.card1.pack(pady=10)

        self.card1.pack_propagate(False)

        self.lbl_foto1 = ctk.CTkLabel(
            self.card1,
            text="Adicionar\nFoto",
            width=150,
            height=150,
            fg_color="gray20",
            corner_radius=10
        )
        self.lbl_foto1.pack(pady=5)

        ctk.CTkLabel(self.card1, text="Jhonatn santos ", font=("Arial", 20, "bold")).pack()
        ctk.CTkLabel(self.card1, text="saldo: R$ 0,00").pack()
        ctk.CTkLabel(self.card1, text="vale: R$ 0,00").pack()
        ctk.CTkLabel(self.card1, text="status: oline").pack()

        self.foto_perfil = ctk.CTkButton(
            self.card1,
            text="Alterar Foto",
            command=self.escolher_foto
        )
        self.foto_perfil.pack(pady=15)

        # Recarregar foto
        if hasattr(self, "caminho_foto"):
            img = Image.open(self.caminho_foto).resize((120, 120))

            self.foto = ctk.CTkImage(
                light_image=img,
                dark_image=img,
                size=(120, 120)
            )

            self.foto_img = self.foto

            self.lbl_foto1.configure(image=self.foto_img, text="")
            self.lbl_foto1.image = self.foto_img

    def criar_servico(self):
        ctk.CTkLabel(
        self.frame_servico,
        text="Adicionar Serviço",
        font=("Arial", 30, "bold")
    ).pack(pady=20)
        
        entryn_nome  = ctk.CTkEntry(
            self.frame_servico,
            placeholder_text="nome",
            width=200,
            height=30,
            corner_radius=300
            )
        entryn_nome.pack(pady=10)

        valor_serviço  = ctk.CTkEntry(
            self.frame_servico,
            placeholder_text="serviço",
            width=200,
            height=30,
            corner_radius=300
            )
        valor_serviço.pack(pady=10)

        forma_pagamento  = ctk.CTkEntry(
            self.frame_servico,
            placeholder_text="forma de pagamento",
            width=200,
            height=30,
            corner_radius=300
            )
        forma_pagamento.pack(pady=10)

        ctk.CTkLabel(
        self.frame_servico,
        text="Descrição",
        font=("Arial",15, "bold")
        ).pack(pady=5)


        descriçao= ctk.CTkTextbox(self.frame_servico, 
        width=200, height=80,
        corner_radius=10,

        )
        descriçao.pack()
       

        ctk.CTkButton(self.frame_servico, 
        text="adicionar", border_width=2,
        border_color="#0A0A0A",
        text_color="#0A0A0A", 
        fg_color="#0E36A3").pack(pady=20)



    def criar_vale(self):
        ctk.CTkLabel(
        self.frame_vale,
        text="Fazer Vale",
        font=("Arial", 30, "bold")
    ).pack(pady=20)
    

    def escolher_foto(self):

        caminho = filedialog.askopenfilename(
            title="Escolha uma foto",
            filetypes=[("Imagens", "*.png *.jpg *.jpeg")]
        )

        if caminho:
            img = Image.open(caminho)
            img = img.resize((120, 120))

            self.foto = ctk.CTkImage(
                light_image=img,
                dark_image=img,
                size=(120, 120)
            )

            # 🔥 IMPORTANTE: uma única linha de update
            self.lbl_foto1.configure(image=self.foto, text="")

            # 🔥 mantém referência (obrigatório)
            self.lbl_foto1.image = self.foto

            # salva caminho
            self.caminho_foto = caminho
App()   

```
