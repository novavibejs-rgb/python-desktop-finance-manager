```python
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from matplotlib.figure import Figure
import customtkinter as ctk
from datetime import datetime

class App:

    def __init__(self):
        self.root = ctk.CTk()
        self.root.title("Sistema Financeiro",)
        self.root.geometry("800x520")
        self.root.resizable(False, False)
        self.root.configure(fg_color="#0E36A3")
    

        # ===== CONFIG GRID PRINCIPAL =====
        self.root.grid_rowconfigure(0, weight=0)  # topo
        self.root.grid_rowconfigure(1, weight=1)  # conteúdo

        self.root.grid_columnconfigure(0, minsize=200)
        self.root.grid_columnconfigure(1, weight=1)  # dashboard

        # ===== TOP BAR (RELÓGIO) =====
        self.top_frame = ctk.CTkFrame(self.root, fg_color="#686666")
        self.top_frame.grid(row=0, column=0, columnspan=2, sticky="ew", padx=10, pady=10)

        self.relogio_label = ctk.CTkLabel(
            self.top_frame,
            text="relogio",
            font=("Segoe UI Emoji", 24,),
            text_color="#F8F5F5"
        )
        self.relogio_label.pack(pady=10)

        # ===== MENU LATERAL =====
        self.menu_frame = ctk.CTkFrame(self.root, fg_color="#686666",width=250)
        self.menu_frame.grid(row=1, column=0, sticky="nsew", padx=10, pady=10)
       
        ctk.CTkLabel(
            self.menu_frame,
            text="MENU",
            font=("Arial", 30, "bold"),
            text_color="#0A0404"
        ).pack(pady=5)

        self.menu_frame.grid_propagate(False)

        #buttons do menu
        ctk.CTkButton(self.menu_frame, text="Inicio",text_color="#0D0E0D",fg_color="#0E36A3",command=self.mostrar_inicio).pack(pady=20)
        ctk.CTkButton(self.menu_frame, text="Perfil", text_color="#0A0A0A", fg_color="#0E36A3",command=self.mostrar_perfil).pack(pady=20)
        ctk.CTkButton(self.menu_frame, text="Adicionar serviço", text_color="#0A0A0A", fg_color="#0E36A3", command=self.mostrar_adicionar_servico).pack(pady=20)
        ctk.CTkButton(self.menu_frame, text="Fazer vale", text_color="#0A0A0A", fg_color="#0E36A3", command=self.mostrar_fazer_vale).pack(pady=20)

        # ===== CONTEÚDO PRINCIPAL =====
        self.content_frame = ctk.CTkFrame(self.root,)
        self.content_frame.grid(row=1,column=1,sticky="nsew",padx=10,pady=10)
        
        
        ## Iniciar com a tela de início
        self.mostrar_inicio()
        self.atualizar_relogio()

        self.root.mainloop()

    # ===== RELÓGIO =====
    def atualizar_relogio(self):
        agora = datetime.now().strftime(" %d/%m/%Y  |  %H:%M:%S")
        self.relogio_label.configure(text=agora)
        self.root.after(1000, self.atualizar_relogio)

    def mostrar_adicionar_servico(self):
        self.content_frame.configure(fg_color="#B8B8EB")
        self.limpar_content()
    

    def mostrar_fazer_vale(self):
        self.content_frame.configure(fg_color="#B8B8EB")
        self.limpar_content()
        

    def mostrar_perfil(self):
        self.content_frame.configure(fg_color="#348AEC")
        self.abrir_perfil()
        
    
    def mostrar_inicio(self):    
        self.content_frame.configure(fg_color="#686666")
        self.abrir_inicio()


    def limpar_content(self):
        for widget in self.content_frame.winfo_children():
            widget.destroy()


    def abrir_inicio(self):
        #iniciio
        self.limpar_content()

        self.rd_pe = ctk.CTkFrame(self.content_frame, fg_color="#20B960", height=100)
        self.rd_pe.grid(row=0, column=2, columnspan=2, sticky="ew",padx=10, pady=(250,10))
        
        # Saldo Total
        ctk.CTkLabel(
            self.rd_pe,
            text="💰 Saldo Total",
            font=("Arial", 25, "bold"),
            text_color="#0D0C0E"
        ).grid( row=0, column=0, pady=10, padx=10)

        self.lbl_saldo_total = ctk.CTkLabel(
            self.rd_pe,
            text="R$ 1.800,00",
            font=("Arial", 25 ,"bold"),
            text_color="#FCF8F8"
        )
        self.lbl_saldo_total.grid( row=1, column=0, pady=10, padx=5)
         
         # Reserva
        ctk.CTkLabel(
            self.rd_pe,
            text="📥 Reserva",
            font=("Arial", 25,"bold"),
            text_color="#0D0C0E"
        ).grid(row=0, column=1, pady=10, padx=10)

        self.lbl_reserva = ctk.CTkLabel(
            self.rd_pe,
            text="R$ 0,00",
            font=("Arial", 25 ,"bold"),
            text_color="#FCF8F8"
        )

        self.lbl_reserva.grid( row=1, column=1, pady=10, padx=5)

          # Lucro Líquido
        ctk.CTkLabel(
            self.rd_pe,
            text="📈 Lucro Líquido",
            font=("Arial", 25,"bold"),
            text_color="#0D0C0E"
        ).grid( row=0, column=2, pady=10, padx=10)

        self.lbl_lucro = ctk.CTkLabel(
            self.rd_pe,
            text="R$ 0,00",
            font=("Arial", 25 ,"bold"),
            text_color="#FCF8F8"
        )

        self.lbl_lucro.grid( row=1, column=2, pady=10, padx=5)

        self.fig = Figure(figsize=(5, 3), dpi=100,)
        self.ax = self.fig.add_subplot(111)
        self.ax.set_title("Desempenho Mensal")


        # dados de teste
        self.ax.plot([1, 2, 3, 4], [10, 25, 15, 30])

        # cria canvas dentro do content_frame
        self.canvas = FigureCanvasTkAgg(
            self.fig,
            master=self.content_frame
        )

        self.canvas.draw()

        self.canvas.get_tk_widget().grid(
        row=0,
        column=3,
        sticky="nsew",
        padx=(10,20),
        pady=(10,200)
        )
        

    def abrir_perfil(self):
        self.limpar_content()
        ctk.CTkLabel(
            self.content_frame,
            text="Perfil do Usuário",
            font=("Arial", 30, "bold"),
            text_color="#0D0C0E"
        ).grid(row=0, column=0, pady=10, padx=170)

    def abrir_adicionar_servico(self):

        pass


    def abrir_fazer_vale(self):
      
        self.limpar_content()
       

App()   

```

```python
dados_semanais = {
    "Semana 1": {"ganhos": 1200, "gastos": 800},
    "Semana 2": {"ganhos": 900, "gastos": 1100},
    "Semana 3": {"ganhos": 1500, "gastos": 1500},
    "Semana 4": {"ganhos": 2000, "gastos": 1200},
}


semanas = []
resultado = []

for semana, valores in dados_semanais.items():
    lucro = valores["ganhos"] - valores["gastos"]
    semanas.append(semana)
    resultado.append(lucro)


import matplotlib.pyplot as plt

cores = ["green" if r > 0 else "red" if r < 0 else "gray" for r in resultado]

plt.bar(semanas, resultado, color=cores)
plt.axhline(0, color="black")
plt.title("Resultado Semanal da Empresa")
plt.ylabel("Lucro / Prejuízo")
plt.show()
```
