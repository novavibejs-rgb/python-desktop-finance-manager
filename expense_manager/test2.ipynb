```python
import customtkinter as ctk


class SistemaFinanceiro:
    def __init__(self):

        # Janela principal
        self.root = ctk.CTk()
        self.root.title("Sistema Financeiro 🚀")
        self.root.geometry("1000x600")

        # Tema
        ctk.set_appearance_mode("dark")
        ctk.set_default_color_theme("blue")

        # Criar interface
        self.criar_menu()
        self.criar_dashboard()

    def criar_menu(self):

        self.menu_frame = ctk.CTkFrame(
            self.root,
            height=50
        )

        self.menu_frame.pack(
            fill="x",
            padx=10,
            pady=10
        )

        ctk.CTkButton(
            self.menu_frame,
            text="Início",
            width=150
        ).pack(side="left", padx=10)

        ctk.CTkButton(
            self.menu_frame,
            text="Serviços",
            width=150
        ).pack(side="left", padx=10)

        ctk.CTkButton(
            self.menu_frame,
            text="Vales",
            width=150
        ).pack(side="left", padx=10)

        ctk.CTkButton(
            self.menu_frame,
            text="Relatórios",
            width=150
        ).pack(side="left", padx=10)

    def criar_dashboard(self):

        # Frame principal
        self.dashboard_frame = ctk.CTkFrame(
            self.root
        )

        self.dashboard_frame.pack(
            fill="both",
            expand=True,
            padx=10,
            pady=(0, 10)
        )

        # Configuração das colunas
        self.dashboard_frame.grid_columnconfigure(
            0,
            weight=1
        )

        self.dashboard_frame.grid_columnconfigure(
            1,
            weight=1
        )

        # Painel financeiro
        self.financeiro_frame = ctk.CTkFrame(
            self.dashboard_frame
        )

        self.financeiro_frame.grid(
            row=0,
            column=0,
            sticky="nsew",
            padx=10,
            pady=10
        )

        # Título
        ctk.CTkLabel(
            self.financeiro_frame,
            text="Resumo Financeiro",
            font=("Arial", 24, "bold")
        ).pack(pady=20)

        # Saldo Total
        ctk.CTkLabel(
            self.financeiro_frame,
            text="💰 Saldo Total",
            font=("Arial", 20, "bold")
        ).pack(pady=(20, 5))

        self.lbl_saldo_total = ctk.CTkLabel(
            self.financeiro_frame,
            text="R$ 1.800,00",
            font=("Arial", 20)
        )

        self.lbl_saldo_total.pack()

        # Reserva
        ctk.CTkLabel(
            self.financeiro_frame,
            text="📥 Reserva",
            font=("Arial", 18)
        ).pack(pady=(20, 5))

        self.lbl_reserva = ctk.CTkLabel(
            self.financeiro_frame,
            text="R$ 0,00",
            font=("Arial", 20)
        )

        self.lbl_reserva.pack()

        # Lucro Líquido
        ctk.CTkLabel(
            self.financeiro_frame,
            text="📈 Lucro Líquido",
            font=("Arial", 18)
        ).pack(pady=(20, 5))

        self.lbl_lucro = ctk.CTkLabel(
            self.financeiro_frame,
            text="R$ 0,00",
            font=("Arial", 20)
        )

        self.lbl_lucro.pack()

    def executar(self):
            self.root.mainloop()


if __name__ == "__main__":
    app = SistemaFinanceiro()
    app.executar()
```
